```python
df = df.rename(columns={'[General Properties/Log(D) at pH=7.4] Predictions': 'Log(D)'})
```

```python
metabolism_cols = [col for col in df.columns if col.startswith('[Metabolism/')]
metabolism_cols = ['Name'] + metabolism_cols
```

```python
meta = df[metabolism_cols]
```

```python
meta
```

```python
def clean_metabolism_col(col):
    # solo procesa columnas Metabolism
    if col.startswith("[Metabolism/"):
        col = col.replace("[Metabolism/", "")  # remove prefix
        col = col.replace("]", "")             # remove final bracket
        col = col.replace(" Probability", "")  # remove probability suffix
    return col

meta = meta.copy()
meta.columns = [clean_metabolism_col(col) for col in meta.columns]
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

## 🛠️ Data Preparation: Select Numerical Columns

numerical_cols = meta.select_dtypes(include=['number']).columns

if numerical_cols.empty:
    print("No numerical columns found in the dataset.")
else:
    sns.set_style("whitegrid")

    num_plots = len(numerical_cols)
    n_cols = 2
    n_rows = (num_plots + n_cols - 1) // n_cols

    fig, axes = plt.subplots(
        nrows=n_rows, 
        ncols=n_cols, 
        figsize=(14, 5 * n_rows)
    )

    axes = axes.flatten()

    for i, column in enumerate(numerical_cols):

        # --- Boxplot (distribution)
        sns.boxplot(y=meta[column], ax=axes[i], color='lightgray')

        # --- Add points colored by compound name (gives legend)
        sns.stripplot(
            y=meta[column],
            x=[""] * len(meta),     # dummy x for vertical layout
            hue=meta["Name"],       # 🎯 add compound names as legend
            dodge=True,
            jitter=True,
            size=5,
            ax=axes[i]
        )

        axes[i].set_title(f'Boxplot of {column}', fontsize=14)
        axes[i].set_ylabel(column, fontsize=12)

        # move legend OUTSIDE the subplot
        axes[i].legend(
            title="Compounds",
            bbox_to_anchor=(1.05, 1),
            loc='upper left',
            borderaxespad=0.
        )

    # Remove empty subplots
    for j in range(num_plots, len(axes)):
        fig.delaxes(axes[j])

    plt.tight_layout()
    plt.savefig("met_boxplots.png", dpi=300, bbox_inches='tight')
    plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# --- IMPORTANT: Assume 'meta' DataFrame is loaded and ready ---
# Include your data filtering logic here, if needed (retaining the 0.5 threshold logic):
# numerical_cols_to_plot = [...]

# --- Example Data Setup (for completeness, based on previous context) ---
# Replace this with your actual data loading and filtering
# meta = pd.DataFrame(...) 
# numerical_cols_to_plot = [...]
numerical_cols_to_plot = [
    'CYP 1A2_substrate', 
    'CYP 2C19_substrate', 
    'CYP 3A4 Inhibitor',
    'CYP 3A4 Substrate',
]
# ----------------------------------------------------------------------


if not numerical_cols_to_plot:
    print("No numerical columns found to plot.")
else:
    ## 📊 Generating Improved Boxplots

    sns.set_style("whitegrid")

    num_plots = len(numerical_cols_to_plot)
    
    # Using 2 columns for the plots
    n_cols = 2 
    n_rows = (num_plots + n_cols - 1) // n_cols 

    # --- ADJUSTED FIGURE SIZE: Create vertical space for the legend above or below ---
    # We'll use extra height (1.5 * n_rows) to reserve space for the legend *above* the plots
    fig, axes = plt.subplots(
        nrows=n_rows, 
        ncols=n_cols, 
        figsize=(14, 7.5 * n_rows) # Increased height per row to 7.5
    )

    # Flatten the axes array for easy iteration
    try:
        axes = axes.flatten()
    except AttributeError:
        axes = [axes] # Handle case where only 1 subplot is created

    # Containers to capture the plotting elements for the unified legend
    legend_handles = []
    legend_labels = []

    for i, column in enumerate(numerical_cols_to_plot):

        # --- Boxplot (distribution)
        sns.boxplot(y=meta[column], ax=axes[i], color='lightgray', linewidth=1)

        # --- Stripplot for individual points
        # The 'hue' parameter generates the colors needed for the legend
        stripplot = sns.stripplot(
            y=meta[column],
            x=[""] * len(meta),       # dummy x for vertical layout
            hue=meta["Name"],         # compound names as legend source
            dodge=True,
            jitter=0.2,               
            size=6,
            ax=axes[i],
            alpha=0.7                 
        )

        axes[i].set_title(f'Boxplot of {column}', fontsize=14)
        axes[i].set_ylabel(column, fontsize=12)
        axes[i].set_xlabel("") # Remove the dummy x-label

        # --- CAPTURE LEGEND DATA FROM THE FIRST PLOT ---
        # We only need to grab the handles and labels once to create the figure legend
        if i == 0: 
            h, l = stripplot.get_legend_handles_labels()
            legend_handles = h
            legend_labels = l
            
        # --- REMOVE IN-PLOT LEGEND ---
        axes[i].get_legend().remove() 

    # Remove empty subplots
    for j in range(num_plots, len(axes)):
        fig.delaxes(axes[j])

    # --- NEW LEGEND PLACEMENT LOGIC ---
    # Place the captured legend centrally above the subplot grid
    fig.legend(
        legend_handles, 
        legend_labels,
        title="Compounds",
        loc="upper center",      # Centered horizontally at the top
        bbox_to_anchor=(0.5, 0.98), # Adjust position vertically (0.98 is near the top edge)
        ncol=4,                  # Adjust number of columns to make it wide/compact
        fontsize='small',
        title_fontsize='medium',
        markerscale=1.2
    )

    # Adjust the layout, reserving space at the top for the figure legend
    # rect=[left, bottom, right, top]
    # We set top=0.88 to push the subplots down, ensuring the legend fits above them.
    plt.tight_layout(rect=[0.0, 0.0, 1.0, 0.88]) 

    plt.savefig("met_boxplots_centered_legend.png", dpi=300) 
    plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# --- 1. DEFINICIÓN DE COLUMNAS ---
# Asumimos que los DataFrames 'phy' y 'meta' ya están cargados.

cyp_cols_meta = [
    'CYP 1A2_substrate', 
    'CYP 2C19_substrate', 
    'CYP 3A4 Inhibitor', 
    'CYP 3A4 Substrate'
]

physchem_cols_phy = [
    'MW', 'Vol', 'Dense', 'nHA', 'nHD', 'TPSA',
    'nRot', 'nRing', 'MaxRing', 'nHet',
    'nRig', 'Flex', 'Log(D)'
]

# --- 2. PREPARACIÓN DE DATOS (Asumiendo que 'phy' y 'meta' tienen el mismo índice) ---
try:
    # 2a. Unir las columnas de los dos DataFrames. 
    # Usamos .copy() para evitar SettingWithCopyWarning
    df_cyp = meta[cyp_cols_meta].copy()
    df_physchem = phy[physchem_cols_phy].copy()

    # Concatenar las columnas para tener un solo DataFrame
    df_combined = pd.concat([df_cyp, df_physchem], axis=1)

except KeyError as e:
    # Este bloque te ayudará a identificar exactamente qué columna falta.
    print(f"\nERROR: ¡Aún existe un KeyError!")
    print(f"La columna faltante es: {e}. Por favor, revisa el listado de columnas.")
    print("Asegúrate de que los nombres en 'cyp_cols_meta' coincidan exactamente con las columnas de 'meta'.")
    print("Y que los nombres en 'physchem_cols_phy' coincidan con las columnas de 'phy'.")
    # Si quieres ver las columnas reales para debuggear:
    # print("Columnas reales en meta:", meta.columns.tolist())
    # print("Columnas reales en phy:", phy.columns.tolist())
else:
    # --- 3. CÁLCULO DE LA CORRELACIÓN ---
    corr_matrix = df_combined.corr()

    # 4. EXTRAER LA SUB-MATRIZ DE INTERÉS
    # Filas: Las columnas CYP (del DataFrame 'meta')
    # Columnas: Las columnas PhysChem (del DataFrame 'phy')
    correlation_cyp_physchem = corr_matrix.loc[cyp_cols_meta, physchem_cols_phy]

    # --- 5. GENERACIÓN DEL MAPA DE CALOR ---
    plt.figure(figsize=(15, 6)) # Ajustado para 14 columnas (ancho) x 4 filas (alto)
    
    sns.heatmap(
        correlation_cyp_physchem,
        annot=True,                 # Muestra el valor de correlación
        fmt=".2f",                  # Formato con 2 decimales
        cmap='coolwarm',            # Usa un esquema de color azul-rojo
        linewidths=.5,              # Espacio entre celdas
        cbar_kws={'label': 'Coeficiente de Correlación de Pearson'},
        vmin=-1, vmax=1             # Asegura que la escala vaya de -1 a 1
    )
    
    plt.title('Correlación entre Actividades CYP y Propiedades Fisicoquímicas', fontsize=16)
    plt.ylabel('Actividades CYP', fontsize=12)
    plt.xlabel('Propiedades Fisicoquímicas', fontsize=12)
    plt.yticks(rotation=0) # Mantiene las etiquetas de las filas horizontales
    
    plt.tight_layout()
    plt.savefig("correlacion_heatmap.png", dpi=300)
    plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# ============================================================
# 1. PREPARACION DE DATOS
# ============================================================

# Columnas CYP provenientes del DataFrame 'meta'
cyp_cols_meta = [
    'CYP 1A2_substrate', 
    'CYP 2C19_substrate', 
    'CYP 3A4 Inhibitor', 
    'CYP 3A4 Substrate'
]

# Columnas fisicoquímicas provenientes del DataFrame 'phy'
physchem_cols_phy = [
    'MW', 'Vol', 'Dense', 'nHA', 'nHD', 'TPSA',
    'nRot', 'nRing', 'MaxRing', 'nHet',
    'nRig', 'Flex', 'Log(D)'
]

try:
    # Tomamos columnas de cada DF
    df_cyp = meta[cyp_cols_meta].copy()
    df_physchem = phy[physchem_cols_phy].copy()

    # Fusionamos
    df_combined = pd.concat([df_cyp, df_physchem], axis=1)

except KeyError as e:
    print(f"\nERROR: Columna faltante → {e}")
    exit()


# ============================================================
# 2. HEATMAP DE CORRELACIONES
# ============================================================

corr_matrix = df_combined.corr()
correlation_cyp_physchem = corr_matrix.loc[cyp_cols_meta, physchem_cols_phy]

plt.figure(figsize=(15, 6))
sns.heatmap(
    correlation_cyp_physchem,
    annot=True,
    fmt=".2f",
    cmap='coolwarm',
    linewidths=.5,
    cbar_kws={'label': 'Pearson r'},
    vmin=-1, vmax=1
)
plt.title('Correlación entre Actividades CYP y Propiedades Fisicoquímicas')
plt.ylabel('Actividades CYP')
plt.xlabel('Propiedades Fisicoquímicas')
plt.tight_layout()
plt.show()


# ============================================================
# 3. SCATTERPLOTS CON LÍNEA DE REGRESIÓN
# ============================================================

pairs_to_plot = [
    ('CYP 3A4 Inhibitor', 'Log(D)', 0.86),
    ('CYP 1A2_substrate', 'MaxRing', 0.84)
]

# Comprobamos que las columnas existen
try:
    df_plot = df_combined[
        ['CYP 3A4 Inhibitor', 'Log(D)', 'CYP 1A2_substrate', 'MaxRing']
    ].copy()
except KeyError as e:
    print(f"\nERROR en scatterplot: columna faltante → {e}")
    exit()

# Figura con 2 gráficas
fig, axes = plt.subplots(1, 2, figsize=(16, 7))
plt.style.use('seaborn-v0_8-whitegrid')

for i, (cyp_col, physchem_col, r_value) in enumerate(pairs_to_plot):
    ax = axes[i]
    
    # Scatterplot
    sns.scatterplot(
        x=df_plot[physchem_col],
        y=df_plot[cyp_col],
        ax=ax,
        color='#1f77b4',
        s=80,
        alpha=0.7
    )
    
    # Línea de regresión
    sns.regplot(
        x=df_plot[physchem_col],
        y=df_plot[cyp_col],
        ax=ax,
        scatter=False,
        color='red',
        line_kws={'linestyle': '--'}
    )
    
    # Título
    ax.set_title(
        f'{cyp_col} vs. {physchem_col}\n(r = {r_value:.2f})',
        fontsize=16
    )
    ax.set_xlabel(physchem_col, fontsize=14)
    ax.set_ylabel(cyp_col, fontsize=14)

plt.suptitle('Gráficos de Dispersión: Correlaciones CYP Más Fuertes', fontsize=18)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# ============================================================
# PARAMETERS
# ============================================================

N_TOP = 5  # Number of top correlations to plot
SAVE_PLOTS = False  # Set True to save plots as PNG

# Column lists (adjust to your actual DataFrames)
cyp_cols = ['CYP 1A2_substrate', 'CYP 2C19_substrate', 'CYP 3A4 Inhibitor', 'CYP 3A4 Substrate']
physchem_cols = ['MW', 'Vol', 'Dense', 'nHA', 'nHD', 'TPSA',
                 'nRot', 'nRing', 'MaxRing', 'nHet', 'nRig', 'Flex', 'Log(D)']

# df_combined is assumed to be your merged DataFrame with both CYP and physchem columns
# df_combined = pd.concat([meta[cyp_cols], phy[physchem_cols]], axis=1)


# ============================================================
# 1. COMPUTE CORRELATIONS
# ============================================================

corr_matrix = df_combined.corr()
# Select only CYP vs PhysChem correlations
corr_cyp_physchem = corr_matrix.loc[cyp_cols, physchem_cols]

# Flatten to a long-form DataFrame
corr_long = corr_cyp_physchem.reset_index().melt(
    id_vars='index', var_name='PhysChem', value_name='Correlation'
)
corr_long.rename(columns={'index': 'CYP'}, inplace=True)

# Take absolute value for ranking
corr_long['AbsCorr'] = corr_long['Correlation'].abs()

# Select top N correlations
top_corrs = corr_long.sort_values('AbsCorr', ascending=False).head(N_TOP)
print("Top correlations:\n", top_corrs[['CYP', 'PhysChem', 'Correlation']])


# ============================================================
# 2. PLOT SCATTERPLOTS FOR TOP CORRELATIONS
# ============================================================

for idx, row in top_corrs.iterrows():
    cyp_col = row['CYP']
    physchem_col = row['PhysChem']
    r_value = row['Correlation']

    plt.figure(figsize=(7, 5))
    
    # Scatterplot
    sns.scatterplot(
        x=df_combined[physchem_col],
        y=df_combined[cyp_col],
        color='#1f77b4',
        s=80,
        alpha=0.7
    )

    # Regression line
    sns.regplot(
        x=df_combined[physchem_col],
        y=df_combined[cyp_col],
        scatter=False,
        color='red',
        line_kws={'linestyle': '--'}
    )

    # Titles and labels
    plt.title(f'{cyp_col} vs {physchem_col}\nCorrelation r = {r_value:.2f}', fontsize=14)
    plt.xlabel(physchem_col, fontsize=12)
    plt.ylabel(cyp_col, fontsize=12)
    plt.tight_layout()

    if SAVE_PLOTS:
        filename = f'scatter_{cyp_col}_{physchem_col}.png'.replace(" ", "_")
        plt.savefig(filename, dpi=300)

    plt.show()
```
