```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import os

# ==============================
# Read data
# ==============================
df = pd.read_excel("cuasinoides.xlsx", sheet_name="pkcsm")

# ==============================
# Define Excretion variables
# ==============================
excretion_cols = [
    "[Excretion/Clearance] Predictions",
    "[Excretion/Organic Cation Transporter 2] Probability",
]

df_excretion = df.loc[:, excretion_cols]

# ==============================
# Clean column names
# ==============================
rename_map = {
    "[Excretion/Clearance] Predictions": "Clearance",
    "[Excretion/Organic Cation Transporter 2] Probability": "OCT2 Probability"
}

df_excretion = df_excretion.rename(columns=rename_map)

# ==============================
# Output directory
# ==============================
OUTPUT_DIR = "4.Excrecin"
os.makedirs(OUTPUT_DIR, exist_ok=True)

# ==============================
# Plotting style
# ==============================
plt.style.use("seaborn-v0_8-whitegrid")
sns.set_palette("husl")

# ==============================
# Boxplot function (consistent with your EDA)
# ==============================
def create_boxplot(ax, data, title, ylabel, color):
    data = data.dropna()

    boxprops = dict(facecolor=color, alpha=0.7, linewidth=1.5)
    whiskerprops = dict(color=color, linewidth=1.5)
    capprops = dict(color=color, linewidth=1.5)
    medianprops = dict(color="black", linewidth=2)

    ax.boxplot(
        data,
        patch_artist=True,
        boxprops=boxprops,
        whiskerprops=whiskerprops,
        capprops=capprops,
        medianprops=medianprops,
        widths=0.6
    )

    # Individual compounds (jitter)
    x = np.random.normal(1, 0.04, size=len(data))
    ax.scatter(
        x,
        data,
        alpha=0.6,
        s=30,
        color=color,
        edgecolors="white",
        linewidth=0.5
    )

    ax.set_title(title, fontsize=14, fontweight="bold", pad=12)
    ax.set_ylabel(ylabel, fontsize=12, fontweight="bold")
    ax.set_xticks([])
    ax.grid(True, alpha=0.3, linestyle="--")
    ax.set_axisbelow(True)

    # Stats annotation
    q1 = data.quantile(0.25)
    q3 = data.quantile(0.75)

    stats_text = (
        f"n = {len(data)}\n"
        f"Median: {data.median():.3f}\n"
        f"Mean: {data.mean():.3f}\n"
        f"IQR: {(q3 - q1):.3f}"
    )

    ax.text(
        0.02, 0.98,
        stats_text,
        transform=ax.transAxes,
        fontsize=10,
        verticalalignment="top",
        bbox=dict(boxstyle="round", facecolor="white", alpha=0.85),
        fontfamily="monospace"
    )

# ==============================
# Create boxplots
# ==============================
fig, axes = plt.subplots(1, 2, figsize=(12, 5))

palette = sns.color_palette()

create_boxplot(
    axes[0],
    df_excretion["Clearance"],
    "Clearance Predictions",
    "Clearance Value",
    palette[0]
)

create_boxplot(
    axes[1],
    df_excretion["OCT2 Probability"],
    "Organic Cation Transporter 2",
    "Probability",
    palette[1]
)

# ==============================
# Global title & save
# ==============================
plt.suptitle(
    "Boxplot Analysis of Excretion Parameters",
    fontsize=16,
    fontweight="bold",
    y=1.05
)

plt.tight_layout()
plt.savefig(
    os.path.join(OUTPUT_DIR, "excretion_boxplots.png"),
    dpi=350,
    bbox_inches="tight"
)
plt.show()
```

```python
import pandas as pd

# ==============================
# Read data
# ==============================
df = pd.read_excel("cuasinoides.xlsx", sheet_name="pkcsm")

# ==============================
# Extract Half-Life variables
# ==============================
df_half_life = df.loc[:, [
    "[Excretion/Half-Life of Drug] Probability",
    "[Excretion/Half-Life of Drug] Interpretation"
]].copy()

# ==============================
# Rename for clarity
# ==============================
df_half_life.columns = ["HalfLife_Probability", "HalfLife_Class"]

# ==============================
# Remove rows with missing data
# ==============================
df_half_life = df_half_life.dropna()

print(df_half_life.head())
```

```python
import pandas as pd
import re

# ==============================
# Extract relevant columns
# ==============================
df_half_life = df.loc[:, [
    "[Excretion/Half-Life of Drug] Probability",
    "[Excretion/Half-Life of Drug] Interpretation"
]].copy()

df_half_life.columns = ["HalfLife_Probability", "HalfLife_Interpretation"]

# ==============================
# Parse Interpretation column
# ==============================
def parse_half_life_interpretation(text):
    if pd.isna(text):
        return pd.Series([None, None])

    # Class
    if "< 3" in text:
        hl_class = "Short (<3 h)"
    elif ">= 3" in text:
        hl_class = "Long (3 h)"
    else:
        hl_class = None

    # Confidence tier
    if "Low Confidence" in text:
        confidence = "Low"
    elif "Medium Confidence" in text:
        confidence = "Medium"
    elif "High Confidence" in text:
        confidence = "High"
    else:
        confidence = None

    return pd.Series([hl_class, confidence])

df_half_life[["HalfLife_Class", "HalfLife_ConfidenceTier"]] = (
    df_half_life["HalfLife_Interpretation"]
    .apply(parse_half_life_interpretation)
)

# ==============================
# Drop raw Interpretation if desired
# ==============================
df_half_life = df_half_life.drop(columns="HalfLife_Interpretation")

print(df_half_life.head())
```

```python
plt.figure(figsize=(8.5, 5.5))
# ==============================
# Output directory
# ==============================
OUTPUT_DIR = "4.Excrecin"
os.makedirs(OUTPUT_DIR, exist_ok=True)
# ------------------------------
# Boxplot (group summary)
# ------------------------------
sns.boxplot(
    data=df_half_life,
    x="HalfLife_Class",
    y="HalfLife_Probability",
    palette=sns.color_palette("husl", desat=0.7),
    width=0.55,
    linewidth=1.6,
    showfliers=False,
    boxprops=dict(alpha=0.75),
    medianprops=dict(color="black", linewidth=2.2),
    whiskerprops=dict(linewidth=1.4),
    capprops=dict(linewidth=1.4)
)

# ------------------------------
# Individual compounds
# ------------------------------
sns.stripplot(
    data=df_half_life,
    x="HalfLife_Class",
    y="HalfLife_Probability",
    color="black",
    size=5.5,
    jitter=0.25,
    alpha=0.6,
    edgecolor="white",
    linewidth=0.4
)

# ------------------------------
# Titles and labels
# ------------------------------
plt.title(
    "Half-Life Prediction Probability\nIndividual Compounds by Class",
    fontsize=15,
    fontweight="bold",
    pad=12
)

plt.xlabel("")
plt.ylabel("Predicted Probability", fontsize=12, fontweight="bold")

# ------------------------------
# Aesthetic cleanup
# ------------------------------
sns.despine(trim=True)
plt.grid(axis="y", alpha=0.25, linestyle="--")
plt.tight_layout()
plt.savefig(
    os.path.join(OUTPUT_DIR, "half-life_boxplots.png"),
    dpi=350,
    bbox_inches="tight")

plt.show()
```
