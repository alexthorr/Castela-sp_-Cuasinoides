```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import os

# ==============================
# Read the Excel file
# ==============================
df = pd.read_csv('results_excretion.csv')

# ==============================
# Output directory (already exists)
# ==============================
OUTPUT_DIR = "4.Excrecin"
os.makedirs(OUTPUT_DIR, exist_ok=True)

# ==============================
# Select required columns
# ==============================
selected_columns = [
    '[Excretion/Organic Cation Transporter 2] Probability',
    '[Excretion/Clearance] Predictions',
    '[Excretion/Half-Life of Drug] Probability'
]

df_selected = df[selected_columns]

# ==============================
# Clean column names
# ==============================
clean_names = {
    '[Excretion/Organic Cation Transporter 2] Probability': 'OCT2\nProbability',
    '[Excretion/Clearance] Predictions': 'Clearance\nPredictions',
    '[Excretion/Half-Life of Drug] Probability': 'Half-Life\nProbability'
}

df_clean = df_selected.rename(columns=clean_names)

# ==============================
# Plotting style
# ==============================
plt.style.use('seaborn-v0_8-whitegrid')
sns.set_palette("husl")

fig, axes = plt.subplots(1, 3, figsize=(16, 6))

colors = ['#2E86AB', '#A23B72', '#F18F01']

def create_enhanced_boxplot(ax, data, title, color, ylabel):
    boxprops = dict(facecolor=color, alpha=0.7, linewidth=1.5)
    whiskerprops = dict(color=color, linewidth=1.5)
    capprops = dict(color=color, linewidth=1.5)
    medianprops = dict(color='white', linewidth=2)

    ax.boxplot(
        data.dropna(),
        patch_artist=True,
        boxprops=boxprops,
        whiskerprops=whiskerprops,
        capprops=capprops,
        medianprops=medianprops,
        widths=0.6
    )

    y = data.dropna()
    x = np.random.normal(1, 0.04, size=len(y))
    ax.scatter(
        x, y, alpha=0.6, color=color,
        s=30, edgecolors='white', linewidth=0.5
    )

    ax.set_title(title, fontsize=14, fontweight='bold', pad=15)
    ax.set_ylabel(ylabel, fontsize=12, fontweight='bold')
    ax.grid(True, alpha=0.3, linestyle='--')
    ax.set_axisbelow(True)
    ax.set_xticks([])

    stats = {
        'Median': data.median(),
        'Mean': data.mean(),
        'Q1': data.quantile(0.25),
        'Q3': data.quantile(0.75)
    }

    text_str = (
        f"n = {len(y)}\n"
        f"Median: {stats['Median']:.3f}\n"
        f"Mean: {stats['Mean']:.3f}\n"
        f"IQR: {stats['Q3'] - stats['Q1']:.3f}"
    )

    ax.text(
        0.02, 0.98, text_str,
        transform=ax.transAxes,
        fontsize=10,
        verticalalignment='top',
        bbox=dict(boxstyle='round', facecolor='white', alpha=0.8),
        fontfamily='monospace'
    )

# ==============================
# Create plots
# ==============================
create_enhanced_boxplot(
    axes[0],
    df_clean['OCT2\nProbability'],
    'Organic Cation Transporter 2\nProbability',
    colors[0],
    'Probability'
)

create_enhanced_boxplot(
    axes[1],
    df_clean['Clearance\nPredictions'],
    'Clearance Predictions',
    colors[1],
    'Clearance Value'
)

create_enhanced_boxplot(
    axes[2],
    df_clean['Half-Life\nProbability'],
    'Drug Half-Life\nProbability',
    colors[2],
    'Probability'
)

plt.suptitle(
    'Excretion Parameters Distribution Analysis',
    fontsize=16,
    fontweight='bold',
    y=1.02
)

plt.tight_layout()
plt.subplots_adjust(wspace=0.3)

# ==============================
# Save figure in existing folder
# ==============================
output_path = os.path.join(OUTPUT_DIR, "excretion2.png")
plt.savefig(output_path, dpi=350, bbox_inches='tight')

plt.show()

# ==============================
# Print detailed statistics
# ==============================
print(" EXCRETION PARAMETERS - DETAILED STATISTICS")
print("=" * 55)

for col in df_clean.columns:
    data = df_clean[col].dropna()
    print(f"\n{col.replace(chr(10), ' ')}:")
    print("-" * 30)
    print(f"Sample size (n):    {len(data):>6}")
    print(f"Mean:               {data.mean():>10.4f}")
    print(f"Median:             {data.median():>10.4f}")
    print(f"Standard deviation: {data.std():>10.4f}")
    print(f"Minimum:            {data.min():>10.4f}")
    print(f"25th percentile:    {data.quantile(0.25):>10.4f}")
    print(f"75th percentile:    {data.quantile(0.75):>10.4f}")
    print(f"Maximum:            {data.max():>10.4f}")
    print(f"IQR:                {data.quantile(0.75) - data.quantile(0.25):>10.4f}")
    print(f"Missing values:     {df_clean[col].isna().sum():>6}")

print("=" * 55)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import os

# ==============================
# Read data
# ==============================
df = pd.read_csv("results_excretion.csv")

# ==============================
# Define Excretion variables
# ==============================
excretion_cols = [
    "[Excretion/Clearance] Predictions",
    "[Excretion/Organic Cation Transporter 2] Probability",
]

df_excretion = df.loc[:, excretion_cols]

# ==============================
# Clean column names
# ==============================
rename_map = {
    "[Excretion/Clearance] Predictions": "Clearance",
    "[Excretion/Organic Cation Transporter 2] Probability": "OCT2 Probability"
}

df_excretion = df_excretion.rename(columns=rename_map)

# ==============================
# Output directory
# ==============================
OUTPUT_DIR = "4.Excrecin"
os.makedirs(OUTPUT_DIR, exist_ok=True)

# ==============================
# Plotting style
# ==============================
plt.style.use("seaborn-v0_8-whitegrid")
sns.set_palette("husl")

# ==============================
# Boxplot function (consistent with your EDA)
# ==============================
def create_boxplot(ax, data, title, ylabel, color):
    data = data.dropna()

    boxprops = dict(facecolor=color, alpha=0.7, linewidth=1.5)
    whiskerprops = dict(color=color, linewidth=1.5)
    capprops = dict(color=color, linewidth=1.5)
    medianprops = dict(color="black", linewidth=2)

    ax.boxplot(
        data,
        patch_artist=True,
        boxprops=boxprops,
        whiskerprops=whiskerprops,
        capprops=capprops,
        medianprops=medianprops,
        widths=0.6
    )

    # Individual compounds (jitter)
    x = np.random.normal(1, 0.04, size=len(data))
    ax.scatter(
        x,
        data,
        alpha=0.6,
        s=30,
        color=color,
        edgecolors="white",
        linewidth=0.5
    )

    ax.set_title(title, fontsize=14, fontweight="bold", pad=12)
    ax.set_ylabel(ylabel, fontsize=12, fontweight="bold")
    ax.set_xticks([])
    ax.grid(True, alpha=0.3, linestyle="--")
    ax.set_axisbelow(True)

    # Stats annotation
    q1 = data.quantile(0.25)
    q3 = data.quantile(0.75)

    stats_text = (
        f"n = {len(data)}\n"
        f"Median: {data.median():.3f}\n"
        f"Mean: {data.mean():.3f}\n"
        f"IQR: {(q3 - q1):.3f}"
    )

    ax.text(
        0.02, 0.98,
        stats_text,
        transform=ax.transAxes,
        fontsize=10,
        verticalalignment="top",
        bbox=dict(boxstyle="round", facecolor="white", alpha=0.85),
        fontfamily="monospace"
    )

# ==============================
# Create boxplots
# ==============================
fig, axes = plt.subplots(1, 2, figsize=(12, 5))

palette = sns.color_palette()

create_boxplot(
    axes[0],
    df_excretion["Clearance"],
    "Clearance Predictions",
    "Clearance Value",
    palette[0]
)

create_boxplot(
    axes[1],
    df_excretion["OCT2 Probability"],
    "Organic Cation Transporter 2",
    "Probability",
    palette[1]
)

# ==============================
# Global title & save
# ==============================
plt.suptitle(
    "Boxplot Analysis of Excretion Parameters",
    fontsize=16,
    fontweight="bold",
    y=1.05
)

plt.tight_layout()
plt.savefig(
    os.path.join(OUTPUT_DIR, "excretion_boxplots22.png"),
    dpi=350,
    bbox_inches="tight"
)
plt.show()
```

```python
import pandas as pd
import re

# ==============================
# Extract relevant columns
# ==============================
df_half_life = df.loc[:, [
    "[Excretion/Half-Life of Drug] Probability",
    "[Excretion/Half-Life of Drug] Interpretation"
]].copy()

df_half_life.columns = ["HalfLife_Probability", "HalfLife_Interpretation"]

# ==============================
# Parse Interpretation column
# ==============================
def parse_half_life_interpretation(text):
    if pd.isna(text):
        return pd.Series([None, None])

    # Class
    if "< 3" in text:
        hl_class = "Short (<3 h)"
    elif ">= 3" in text:
        hl_class = "Long (3 h)"
    else:
        hl_class = None

    # Confidence tier
    if "Low Confidence" in text:
        confidence = "Low"
    elif "Medium Confidence" in text:
        confidence = "Medium"
    elif "High Confidence" in text:
        confidence = "High"
    else:
        confidence = None

    return pd.Series([hl_class, confidence])

df_half_life[["HalfLife_Class", "HalfLife_ConfidenceTier"]] = (
    df_half_life["HalfLife_Interpretation"]
    .apply(parse_half_life_interpretation)
)

# ==============================
# Drop raw Interpretation if desired
# ==============================
df_half_life = df_half_life.drop(columns="HalfLife_Interpretation")

print(df_half_life.head())
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import os

# ==============================
# Read the Excel file
# ==============================
df = pd.read_csv('results_excretion.csv')

# ==============================
# Output directory (already exists)
# ==============================
OUTPUT_DIR = "4.Excrecin"
os.makedirs(OUTPUT_DIR, exist_ok=True)

# ==============================
# Select required columns
# ==============================
selected_columns = [
    '[Excretion/Organic Cation Transporter 2] Probability',
    '[Excretion/Clearance] Predictions',
    '[Excretion/Half-Life of Drug] Probability'
]

df_selected = df[selected_columns]

# ==============================
# Clean column names
# ==============================
clean_names = {
    '[Excretion/Organic Cation Transporter 2] Probability': 'OCT2 Probability',
    '[Excretion/Clearance] Predictions': 'Clearance Predictions',
}

df_clean = df_selected.rename(columns=clean_names)

# ==============================
# Plotting style
# ==============================
plt.style.use('seaborn-v0_8-whitegrid')
sns.set_palette("husl")

fig, axes = plt.subplots(1, 2, figsize=(16, 5))

# ==============================
# Histogram plotting function
# ==============================
def create_histogram(ax, data, title, xlabel, color):
    data = data.dropna()

    sns.histplot(
        data,
        bins=20,
        kde=True,
        stat="count",
        ax=ax,
        color=color,
        edgecolor="black",
        alpha=0.7
    )

    ax.set_title(title, fontsize=14, fontweight='bold', pad=12)
    ax.set_xlabel(xlabel, fontsize=12, fontweight='bold')
    ax.set_ylabel("Count", fontsize=12)
    ax.grid(True, alpha=0.3, linestyle='--')
    ax.set_axisbelow(True)

    # Statistical annotations
    mean = data.mean()
    median = data.median()
    std = data.std()

    ax.axvline(mean, color='black', linestyle='--', linewidth=1.5, label='Mean')
    ax.axvline(median, color='red', linestyle=':', linewidth=1.5, label='Median')

    ax.legend(fontsize=9)

    text_str = (
        f"n = {len(data)}\n"
        f"Mean: {mean:.3f}\n"
        f"Median: {median:.3f}\n"
        f"SD: {std:.3f}"
    )

    ax.text(
        0.97, 0.97, text_str,
        transform=ax.transAxes,
        fontsize=10,
        verticalalignment='top',
        horizontalalignment='right',
        bbox=dict(boxstyle='round', facecolor='white', alpha=0.85),
        fontfamily='monospace'
    )

# ==============================
# Create histograms
# ==============================
create_histogram(
    axes[0],
    df_clean['OCT2 Probability'],
    'Organic Cation Transporter 2',
    'Probability',
    sns.color_palette()[0]
)

create_histogram(
    axes[1],
    df_clean['Clearance Predictions'],
    'Clearance Predictions',
    'Clearance Value',
    sns.color_palette()[1]
)

# ==============================
# Global title and layout
# ==============================
plt.suptitle(
    'Excretion Parameters  Histogram Analysis',
    fontsize=16,
    fontweight='bold',
    y=1.05
)

plt.tight_layout()
plt.subplots_adjust(wspace=0.3)

# ==============================
# Save figure
# ==============================
output_path = os.path.join(OUTPUT_DIR, "excretion_histograms2.png")
plt.savefig(output_path, dpi=350, bbox_inches='tight')

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# ==============================
# Read data
# ==============================
df = pd.read_csv("results_excretion.csv")

OUTPUT_DIR = "4.Excrecin"
os.makedirs(OUTPUT_DIR, exist_ok=True)
# ==============================
# Extract Half-Life columns
# ==============================
df_half_life = df[[
    "[Excretion/Half-Life of Drug] Probability",
    "[Excretion/Half-Life of Drug] Interpretation"
]].dropna()

# ==============================
# Clean interpretation  Class
# ==============================
df_half_life["HalfLife_Class"] = (
    df_half_life["[Excretion/Half-Life of Drug] Interpretation"]
    .str.extract(r"(Half-Life\s*[<>]=?\s*\d+hs)")
)

df_half_life = df_half_life.rename(columns={
    "[Excretion/Half-Life of Drug] Probability": "HalfLife_Probability"
})

# ==============================
# Plot style
# ==============================
plt.style.use("seaborn-v0_8-whitegrid")
sns.set_palette("husl")

# ==============================
# Create histogram
# ==============================
plt.figure(figsize=(8.5, 5.5))

sns.histplot(
    data=df_half_life,
    x="HalfLife_Probability",
    hue="HalfLife_Class",
    bins=20,
    kde=True,
    stat="count",
    alpha=0.6,
    edgecolor="black"
)

# ==============================
# Global statistics
# ==============================
mean = df_half_life["HalfLife_Probability"].mean()
median = df_half_life["HalfLife_Probability"].median()
std = df_half_life["HalfLife_Probability"].std()

plt.axvline(mean, color="black", linestyle="--", linewidth=1.6, label="Mean")
plt.axvline(median, color="red", linestyle=":", linewidth=1.6, label="Median")

# ==============================
# Titles & labels
# ==============================
plt.title(
    "Half-Life Prediction Probability\nDistribution by Class",
    fontsize=15,
    fontweight="bold",
    pad=12
)

plt.xlabel("Predicted Probability", fontsize=12, fontweight="bold")
plt.ylabel("Density", fontsize=12)

# ==============================
# Annotation box
# ==============================
text_str = (
    f"n = {len(df_half_life)}\n"
    f"Mean: {mean:.3f}\n"
    f"Median: {median:.3f}\n"
    f"SD: {std:.3f}"
)

plt.text(
    0.97, 0.95,
    text_str,
    transform=plt.gca().transAxes,
    fontsize=10,
    verticalalignment="top",
    horizontalalignment="right",
    bbox=dict(boxstyle="round", facecolor="white", alpha=0.85),
    fontfamily="monospace"
)

# ==============================
# Cleanup
# ==============================
sns.despine(trim=True)
plt.grid(axis="y", alpha=0.3, linestyle="--")
plt.legend(title="Half-Life Class", fontsize=10)
plt.tight_layout()
plt.savefig(
    os.path.join(OUTPUT_DIR, "half-life_HIST2.png"),
    dpi=350,
    bbox_inches="tight")
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
from matplotlib.patches import Patch

# Load data
df = pd.read_csv('results_excretion.csv')

OUTPUT_DIR = "4.Excrecin"
os.makedirs(OUTPUT_DIR, exist_ok=True)

# Create simplified IDs
df['Compound ID'] = [f'Q-{i+1}' for i in range(len(df))]

# Sort by Clearance for the ranked view
df_sorted = df.sort_values(by='[Excretion/Clearance] Predictions', ascending=True)

# Define colors
# Using a specific palette for better aesthetics
color_map = {
    'Long': '#4c72b0', # Seaborn deep blue
    'Short': '#c44e52'   # Seaborn deep red
}

# Apply colors based on the prediction string
colors = df_sorted['[Excretion/Half-Life of Drug] Predictions'].apply(
    lambda x: color_map['Short'] if 'Half-Life < 3hs' in x else color_map['Long']
)

# Set style
sns.set_theme(style="whitegrid", context="paper", font_scale=1.2)

# Create Plot
plt.figure(figsize=(10, 10))
bars = plt.barh(df_sorted['Compound ID'], df_sorted['[Excretion/Clearance] Predictions'], color=colors, edgecolor='black', linewidth=0.5)

# Add value labels to the end of each bar
for bar in bars:
    width = bar.get_width()
    plt.text(
        width + 0.2,       # x-position
        bar.get_y() + bar.get_height()/2, # y-position
        f'{width:.2f}',    # label
        ha='left',         # horizontal alignment
        va='center',       # vertical alignment
        fontsize=10,
        color='black'
    )

# Labels and Title
plt.xlabel('Predicted Total Clearance (ml/min/kg)', fontsize=12, fontweight='bold')
plt.ylabel('Compound ID', fontsize=12, fontweight='bold')
plt.title('Predicted Total Clearance and Half-Life Classification\nfor 24 Quassinoids', fontsize=14, fontweight='bold', pad=20)

# Custom Legend
legend_elements = [
    Patch(facecolor=color_map['Long'], edgecolor='black', linewidth=0.5, label=r'Half-Life $\geq$ 3h'),
    Patch(facecolor=color_map['Short'], edgecolor='black', linewidth=0.5, label=r'Half-Life < 3h')
]
plt.legend(handles=legend_elements, title='Half-Life Prediction', loc='lower right', frameon=True, framealpha=0.9)

# Adjust layout and save
plt.tight_layout()
plt.savefig(
    os.path.join(OUTPUT_DIR, "Ranked Total Clearance with Half-Life Overlay.png"),
    dpi=350,
    bbox_inches="tight")
plt.show()
```
