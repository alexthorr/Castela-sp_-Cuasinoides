```python
import os
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib.ticker import MaxNLocator

# ==============================
# Output directory
# ==============================
OUTPUT_DIR = "5.Metabolismo"
os.makedirs(OUTPUT_DIR, exist_ok=True)

# ==============================
# Plotting style
# ==============================
plt.style.use("seaborn-v0_8-whitegrid")
sns.set_palette("husl")

# ==============================
# CYP column groups (EXACT names)
# ==============================
cyp_inhibitors = [
    "CYP 1A2 Inhibitor",
    "CYP 2C19 Inhibitor",
    "CYP 2C9 Inhibitor",
    "CYP 2D6 Inhibitor",
    "CYP 3A4 Inhibitor"
]

cyp_substrates = [
    "CYP 1A2 substrate",
    "CYP 2C19 substrate",
    "CYP 2D6 Substrate",
    "CYP 3A4 Substrate"
]

# ==============================
# Histogram function
# ==============================
def create_histogram(ax, data, title, color):
    data = data.dropna()

    # Disable KDE for near-degenerate distributions
    use_kde = data.std() > 0.05

    sns.histplot(
        data,
        bins=20,
        kde=use_kde,
        stat="count",
        ax=ax,
        color=color,
        edgecolor="black",
        alpha=0.75
    )

    ax.set_xlim(0, 1)
    ax.set_xlabel("Probability", fontsize=11, fontweight="bold")
    ax.set_ylabel("Count", fontsize=11, fontweight="bold")
    ax.set_title(title, fontsize=13, fontweight="bold")

    ax.yaxis.set_major_locator(MaxNLocator(integer=True))
    ax.grid(True, alpha=0.3, linestyle="--")
    ax.set_axisbelow(True)

    mean = data.mean()
    median = data.median()
    std = data.std()

    ax.axvline(mean, color="black", linestyle="--", linewidth=1.5, label="Mean")
    ax.axvline(median, color="red", linestyle=":", linewidth=1.5, label="Median")
    ax.legend(fontsize=9)

    text_str = (
        f"n = {len(data)}\n"
        f"Mean: {mean:.3f}\n"
        f"Median: {median:.3f}\n"
        f"SD: {std:.3f}"
    )

    ax.text(
        0.97, 0.97,
        text_str,
        transform=ax.transAxes,
        fontsize=10,
        verticalalignment="top",
        horizontalalignment="right",
        bbox=dict(boxstyle="round", facecolor="white", alpha=0.85),
        fontfamily="monospace"
    )

# ==============================
# FIGURE 1  CYP INHIBITORS
# ==============================
ncols = 3
nrows = int(np.ceil(len(cyp_inhibitors) / ncols))

fig, axes = plt.subplots(nrows, ncols, figsize=(16, 5 * nrows))
axes = axes.flatten()

palette = sns.color_palette()

for i, col in enumerate(cyp_inhibitors):
    create_histogram(
        ax=axes[i],
        data=meta[col],
        title=col,
        color=palette[i % len(palette)]
    )

for j in range(len(cyp_inhibitors), len(axes)):
    fig.delaxes(axes[j])

plt.suptitle(
    "CYP Inhibition Probability Distributions",
    fontsize=16,
    fontweight="bold",
    y=1.02
)

plt.tight_layout()
plt.savefig(
    os.path.join(OUTPUT_DIR, "cyp_inhibitors_probability_histograms.png"),
    dpi=350,
    bbox_inches="tight"
)
plt.show()

# ==============================
# FIGURE 2  CYP SUBSTRATES
# ==============================
ncols = 2
nrows = int(np.ceil(len(cyp_substrates) / ncols))

fig, axes = plt.subplots(nrows, ncols, figsize=(14, 5 * nrows))
axes = axes.flatten()

for i, col in enumerate(cyp_substrates):
    create_histogram(
        ax=axes[i],
        data=meta[col],
        title=col,
        color=palette[i % len(palette)]
    )

for j in range(len(cyp_substrates), len(axes)):
    fig.delaxes(axes[j])

plt.suptitle(
    "CYP Substrate Probability Distributions",
    fontsize=16,
    fontweight="bold",
    y=1.02
)

plt.tight_layout()
plt.savefig(
    os.path.join(OUTPUT_DIR, "cyp_substrates_probability_histograms.png"),
    dpi=350,
    bbox_inches="tight"
)
plt.show()
```

```python
import os
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# ==============================
# Output directory
# ==============================
OUTPUT_DIR = "5.Metabolismo"
os.makedirs(OUTPUT_DIR, exist_ok=True)

# ==============================
# CYP substrate variables (EXACT names)
# ==============================
cyp_substrate_columns = [
    "CYP 3A4 Substrate",
    "CYP 2C19 substrate"
]

# All variables are probabilities
variable_types = {col: "probability" for col in cyp_substrate_columns}

# ==============================
# Boxplot function
# ==============================
def create_boxplot(ax, data, title, ylabel, color):
    data = data.dropna()

    boxprops = dict(facecolor=color, alpha=0.7, linewidth=1.5)
    whiskerprops = dict(color=color, linewidth=1.5)
    capprops = dict(color=color, linewidth=1.5)
    medianprops = dict(color="black", linewidth=2)

    ax.boxplot(
        data,
        patch_artist=True,
        boxprops=boxprops,
        whiskerprops=whiskerprops,
        capprops=capprops,
        medianprops=medianprops,
        widths=0.6
    )

    # Jittered points
    x = np.random.normal(1, 0.04, size=len(data))
    ax.scatter(
        x,
        data,
        alpha=0.6,
        s=35,
        color=color,
        edgecolors="white",
        linewidth=0.6
    )

    ax.set_title(title, fontsize=14, fontweight="bold", pad=12)
    ax.set_ylabel(ylabel, fontsize=12, fontweight="bold")
    ax.set_xticks([])
    ax.set_ylim(0, 1)
    ax.grid(True, alpha=0.3, linestyle="--")
    ax.set_axisbelow(True)

    q1 = data.quantile(0.25)
    q3 = data.quantile(0.75)

    text_str = (
        f"n = {len(data)}\n"
        f"Median: {data.median():.3f}\n"
        f"Mean: {data.mean():.3f}\n"
        f"IQR: {(q3 - q1):.3f}"
    )

    ax.text(
        0.02, 0.98,
        text_str,
        transform=ax.transAxes,
        fontsize=10,
        verticalalignment="top",
        bbox=dict(boxstyle="round", facecolor="white", alpha=0.85),
        fontfamily="monospace"
    )

# ==============================
# Create boxplots
# ==============================
ncols = 2
nrows = 1

fig, axes = plt.subplots(nrows, ncols, figsize=(14, 6))
axes = axes.flatten()

palette = sns.color_palette("Set2")

for i, col in enumerate(cyp_substrate_columns):
    create_boxplot(
        ax=axes[i],
        data=meta[col],
        title=col,
        ylabel="Probability",
        color=palette[i % len(palette)]
    )

# ==============================
# Global title & save
# ==============================
plt.suptitle(
    "Boxplot Analysis of CYP Substrate Probabilities",
    fontsize=16,
    fontweight="bold",
    y=1.03
)

plt.tight_layout()
plt.savefig(
    os.path.join(OUTPUT_DIR, "cyp_substrate_probability_boxplots.png"),
    dpi=350,
    bbox_inches="tight"
)
plt.show()
```

```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# =========================================================
# Load descriptor data
# =========================================================
df = pd.read_excel("cuasinoides.xlsx", sheet_name="pkcsm")

# =========================================================
# Output directory
# =========================================================
OUTPUT_DIR = "5.Metabolismo"
os.makedirs(OUTPUT_DIR, exist_ok=True)

# =========================================================
# Plot style
# =========================================================
plt.style.use("seaborn-v0_8-whitegrid")
sns.set_context("notebook")

# =========================================================
# CYP substrate variables (from meta)
# =========================================================
cyp_vars = [
    "CYP 3A4 Substrate",
    "CYP 2C19 substrate"
]

# =========================================================
# Descriptor columns (from df)
# =========================================================
descriptor_cols = [
    "MW",
    "nHA",
    "nHD",
    "[General Properties/Log(P)] Predictions",
    "TPSA",
    "Fsp3",
    "nRing",
    "nHet",
    "nRot"
]

# =========================================================
# Descriptor name cleaning (for labels only)
# =========================================================
def clean_descriptor(col):
    col = col.replace("[General Properties/", "")
    col = col.replace("] Predictions", "")
    return col.strip()

descriptor_labels = [clean_descriptor(col) for col in descriptor_cols]

# =========================================================
# Bivariate plotting function
# =========================================================
def bivariate_plot(ax, x, y, x_label, y_label):
    sns.scatterplot(
        x=x,
        y=y,
        ax=ax,
        s=45,
        alpha=0.75,
        edgecolor="black",
        linewidth=0.4
    )

    # LOWESS smoothing (robust for small n)
    sns.regplot(
        x=x,
        y=y,
        ax=ax,
        scatter=False,
        lowess=True,
        color="red",
        line_kws={"linewidth": 2}
    )

    ax.set_xlabel(x_label, fontsize=11, fontweight="bold")
    ax.set_ylabel(y_label, fontsize=11, fontweight="bold")
    ax.set_ylim(0, 1)
    ax.grid(True, alpha=0.3, linestyle="--")
    ax.set_axisbelow(True)

# =========================================================
# Generate bivariate plots
# =========================================================
for cyp_var in cyp_vars:

    ncols = 3
    nrows = int(np.ceil(len(descriptor_cols) / ncols))

    fig, axes = plt.subplots(
        nrows, ncols,
        figsize=(16, 5 * nrows)
    )
    axes = axes.flatten()

    for i, (desc_col, desc_label) in enumerate(zip(descriptor_cols, descriptor_labels)):
        bivariate_plot(
            ax=axes[i],
            x=df[desc_col],
            y=meta[cyp_var],
            x_label=desc_label,
            y_label=cyp_var
        )

    # Remove unused subplots
    for j in range(len(descriptor_cols), len(axes)):
        fig.delaxes(axes[j])

    plt.suptitle(
        f"Physicochemical Descriptors vs {cyp_var}",
        fontsize=16,
        fontweight="bold",
        y=1.02
    )

    plt.tight_layout()
    plt.savefig(
        os.path.join(
            OUTPUT_DIR,
            f"bivariate_descriptors_vs_{cyp_var.replace(' ', '_')}.png"
        ),
        dpi=350,
        bbox_inches="tight"
    )
    plt.show()

```

```python
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import spearmanr

# =========================================================
# Configuration
# =========================================================
OUTPUT_DIR = "5.Metabolismo"
os.makedirs(OUTPUT_DIR, exist_ok=True)

MAX_PAIRPLOT_VARS = 8
MAX_PAIRPLOT_ROWS = 2000

# =========================================================
# Load descriptor data from Excel
# =========================================================
df_desc = pd.read_excel("cuasinoides.xlsx", sheet_name="pkcsm")

# Rename columns if needed
RENAME_MAP = {
    "[General Properties/Log(P)] Predictions": "LogP",
    
}
df_desc = df_desc.rename(columns={k: v for k, v in RENAME_MAP.items() if k in df_desc.columns})

descriptor_cols = ["MW", "LogP", "TPSA", "Fsp3", "nHA", "nHD", "nRot", "nRing", "nHet"]
descriptor_cols = [c for c in descriptor_cols if c in df_desc.columns]

df_desc = df_desc[descriptor_cols].copy()

# =========================================================
# Load CYP substrate data from meta DataFrame
# =========================================================
cyp_substrate_columns = ["CYP 3A4 Substrate", "CYP 2C19 substrate"]
cyp_substrate_columns = [c for c in cyp_substrate_columns if c in meta.columns]
df_cyp = meta[cyp_substrate_columns].copy()

# =========================================================
# Combine descriptor and CYP data
# =========================================================
df = pd.concat([df_desc.reset_index(drop=True),
                df_cyp.reset_index(drop=True)],
               axis=1)

# =========================================================
# Convert to numeric
# =========================================================
df = df.apply(pd.to_numeric, errors="coerce")
df = df.dropna(how="all")

# =========================================================
# Sanity check
# =========================================================
missing = [c for c in cyp_substrate_columns if c not in df.columns]
if missing:
    raise ValueError(f"Missing required CYP substrate columns: {missing}")

# =========================================================
# Spearman correlation + p-values
# =========================================================
def spearman_corr_pval(df_in):
    cols = df_in.columns
    corr = pd.DataFrame(np.nan, index=cols, columns=cols)
    pval = pd.DataFrame(np.nan, index=cols, columns=cols)

    for a in cols:
        for b in cols:
            mask = df_in[a].notna() & df_in[b].notna()
            if mask.sum() >= 3:
                r, p = spearmanr(df_in.loc[mask, a], df_in.loc[mask, b])
                corr.loc[a, b] = r
                pval.loc[a, b] = p
    return corr, pval

corr_df, pval_df = spearman_corr_pval(df)

# =========================================================
# Heatmap annotations
# =========================================================
def significance_stars(p):
    if pd.isna(p):
        return ""
    if p <= 0.001:
        return "***"
    if p <= 0.01:
        return "**"
    if p <= 0.05:
        return "*"
    return ""

annot = pd.DataFrame("", index=corr_df.index, columns=corr_df.columns)
for i in corr_df.index:
    for j in corr_df.columns:
        if not pd.isna(corr_df.loc[i, j]):
            annot.loc[i, j] = f"{corr_df.loc[i, j]:.2f}{significance_stars(pval_df.loc[i, j])}"

# =========================================================
# Plot heatmap
# =========================================================
sns.set_theme(style="whitegrid", context="notebook")
mask = np.triu(np.ones_like(corr_df, dtype=bool))

fig, ax = plt.subplots(figsize=(max(7, 0.6 * len(corr_df)),
                                max(5, 0.5 * len(corr_df))))

sns.heatmap(
    corr_df,
    mask=mask,
    cmap=sns.diverging_palette(220, 10, as_cmap=True),
    center=0,
    annot=annot,
    fmt="",
    annot_kws={"size": 8},
    linewidths=0.3,
    cbar_kws={"shrink": 0.6},
    vmin=-1,
    vmax=1,
    ax=ax
)

ax.set_title(
    "Spearman Correlation Matrix\n(CYP Substrate vs Descriptors)",
    fontsize=12,
    fontweight="bold"
)

plt.tight_layout()
plt.savefig(os.path.join(OUTPUT_DIR, "spearman_cyp_heatmap.png"), dpi=300, bbox_inches="tight")
plt.show()

# =========================================================
# Pairplot
# =========================================================
vars_for_pairplot = cyp_substrate_columns.copy()
remaining = MAX_PAIRPLOT_VARS - len(vars_for_pairplot)

if remaining > 0:
    avg_abs_corr = (
        corr_df.loc[descriptor_cols, cyp_substrate_columns]
        .abs()
        .mean(axis=1)
        .sort_values(ascending=False)
    )
    vars_for_pairplot += avg_abs_corr.head(remaining).index.tolist()

pairplot_df = df[vars_for_pairplot].dropna(how="all")
if len(pairplot_df) > MAX_PAIRPLOT_ROWS:
    pairplot_df = pairplot_df.sample(MAX_PAIRPLOT_ROWS, random_state=42)

pp = sns.pairplot(pairplot_df, diag_kind="hist", plot_kws={"s": 30, "alpha": 0.7})
pp.fig.suptitle("Bivariate Analysis: CYP Substrate vs Descriptors", fontsize=14, fontweight="bold", y=1.02)
pp.fig.savefig(os.path.join(OUTPUT_DIR, "pairplot_cyp.png"), dpi=300, bbox_inches="tight")
plt.show()

# =========================================================
# Save outputs
# =========================================================
corr_df.to_csv(os.path.join(OUTPUT_DIR, "spearman_cyp_corr.csv"))
pval_df.to_csv(os.path.join(OUTPUT_DIR, "spearman_cyp_pvalues.csv"))
pairplot_df.to_csv(os.path.join(OUTPUT_DIR, "pairplot_cyp_data.csv"), index=False)

print("CYP substrate analysis completed successfully.")

```
