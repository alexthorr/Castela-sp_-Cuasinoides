```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from math import pi

# 1. Load data
file_path = 'cuasinoides.xlsx'
df = pd.read_excel(file_path, sheet_name='ADMET')

# Define columns (EXACTLY as in the figure)
# -------------------------------------------------
inhibitor_cols = [
    'OATP1B1 Inhibitor',
    'OATP1B3 Inhibitor',
    'BCRP Inhibitor',
    'BSEP Inhibitor',
    'MRP1 Inhibitor'
]

pk_cols = [
    'PPB Percentage',
    'Fraction Unbound',
    'VDss (L/kg)',
    'BBB Penetration'
]

all_cols = pk_cols + inhibitor_cols  # PK first, transporters second

# -------------------------------------------------
# Data cleaning
# -------------------------------------------------
df[all_cols] = df[all_cols].apply(pd.to_numeric, errors='coerce')
df = df.dropna(subset=all_cols)

# -------------------------------------------------
# Spearman correlation
# -------------------------------------------------
corr_matrix = df[all_cols].corr(method='spearman')

# -------------------------------------------------
# Clustered heatmap
# -------------------------------------------------
g = sns.clustermap(
    corr_matrix,
    cmap='coolwarm',
    center=0,
    annot=True,
    fmt=".2f",
    linewidths=0.6,
    figsize=(12, 12),
    cbar_kws={'label': 'Spearman correlation'}
)

# -------------------------------------------------
# Proper title placement (IMPORTANT)
# -------------------------------------------------
g.fig.suptitle(
    'Correlation Analysis of Distribution Properties for Quassinoids',
    fontsize=16,
    y=1.03
)

# Improve label readability
plt.setp(g.ax_heatmap.get_xticklabels(), rotation=45, ha='right')
plt.setp(g.ax_heatmap.get_yticklabels(), rotation=0)

# -------------------------------------------------
# Save
# -------------------------------------------------
g.savefig('Figure2_ADMET_correlation_heatmap.png', dpi=300, bbox_inches='tight')
plt.show()

# --- PLOT 2: PHARMACOLOGICAL DISTRIBUTION SCATTER ---
# Visualizes the relationship between VDss, Protein Binding, and BBB Penetration
plt.figure(figsize=(10, 7))
# Use log scale for BBB Penetration as it often spans several orders of magnitude
df['Log_BBB'] = np.log10(df['BBB Penetration'] + 1e-9) 

scatter = sns.scatterplot(
    data=df,
    x='VDss (L/kg)',
    y='PPB Percentage',
    hue='Log_BBB',
    palette='viridis',
    s=120,
    alpha=0.75
)
from scipy.stats import spearmanr
r, p = spearmanr(df['VDss (L/kg)'], df['PPB Percentage'])
plt.text(0.05, 0.95, f"Spearman r = {r:.2f}, p = {p:.2g}",
         transform=plt.gca().transAxes)

plt.title('Distribution Profile: $V_{Dss}$ vs. PPB', fontsize=16)
plt.xlabel('$V_{Dss}$ (L/kg)', fontsize=12)
plt.ylabel('Plasma Protein Binding (%)', fontsize=12)
plt.legend(title='Log(BBB) & % Fraction Unbound', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True, linestyle='--', alpha=0.6)
plt.tight_layout()
plt.savefig('distribution_scatter.png', dpi=300)

# --- PLOT 3: COMPARATIVE RADAR CHART (TOP PK LEADS) ---
# Select top 3 compounds with highest VDss and Fraction Unbound
top_df = df.sort_values(
    by=['VDss (L/kg)', 'Fraction Unbound'],
    ascending=[False, False]
).head(3).reset_index(drop=True)

# Radar chart setup
N = len(inhibitor_cols)
angles = [n / float(N) * 2 * pi for n in range(N)]
angles += angles[:1]

fig, ax = plt.subplots(figsize=(8, 8), subplot_kw=dict(polar=True))

colors = ['#E63946', '#457B9D', '#2A9D8F']

for i, row in top_df.iterrows():
    values = row[inhibitor_cols].values.flatten().tolist()
    values += values[:1]
    
    ax.plot(
        angles, values,
        linewidth=2,
        linestyle='solid',
        label=f'Lead {i+1}',
        color=colors[i]
    )
    ax.fill(angles, values, color=colors[i], alpha=0.08)

# Axis formatting
plt.xticks(angles[:-1], inhibitor_cols, size=10)
ax.set_rlabel_position(0)
plt.yticks(
    [0.25, 0.50, 0.75, 1.0],
    ["0.25", "0.50", "0.75", "1.0"],
    color="grey",
    size=9
)
plt.ylim(0, 1)

# Title and legend
plt.title(
    'Transporter Inhibitory Fingerprints of Top PK Leads',
    size=16,
    y=1.08
)
plt.legend(
    loc='upper center',
    bbox_to_anchor=(0.5, -0.10),
    ncol=3,
    frameon=False
)
plt.tight_layout()
plt.savefig('lead_radar_chart.png', dpi=300)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# -------------------------------------------------
# Load data
# -------------------------------------------------
file_path = 'cuasinoides.xlsx'
df = pd.read_excel(file_path, sheet_name='ADMET')

# Ensure numeric types
pk_cols = [
    'VDss (L/kg)',
    'PPB Percentage',
    'Fraction Unbound',
    'BBB Penetration'
]

df[pk_cols] = df[pk_cols].apply(pd.to_numeric, errors='coerce')
df = df.dropna(subset=pk_cols)

# Log-transform BBB penetration (for visualization only)
df['log_BBB'] = np.log10(df['BBB Penetration'] + 1e-9)

# -------------------------------------------------
# Figure 1: Violin plots with individual points
# -------------------------------------------------
sns.set_style("whitegrid")

fig, axes = plt.subplots(1, 4, figsize=(14, 4), dpi=300)

plot_vars = [
    ('VDss (L/kg)', r'$V_{Dss}$ (L/kg)'),
    ('PPB Percentage', 'Plasma Protein Binding (%)'),
    ('Fraction Unbound', 'Fraction Unbound'),
    ('log_BBB', 'log$_{10}$(BBB Penetration)')
]

for ax, (col, label) in zip(axes, plot_vars):
    sns.violinplot(
        y=df[col],
        ax=ax,
        inner=None,
        color='#D0D0D0',
        linewidth=1
    )
    
    sns.stripplot(
        y=df[col],
        ax=ax,
        color='black',
        size=4,
        jitter=True,
        alpha=0.7
    )
    
    median_val = df[col].median()
    ax.axhline(median_val, color='red', linestyle='--', linewidth=1)
    
    ax.set_title(label, fontsize=11)
    ax.set_xlabel('')
    ax.set_ylabel('')

plt.suptitle(
    'Distribution of Key Pharmacokinetic Endpoints for Cuasinoids',
    fontsize=14,
    y=1.05
)

plt.tight_layout()
plt.savefig('Figure1_PK_distributions.png', dpi=300, bbox_inches='tight')
plt.show()
```
